```python
import numpy as np
import matplotlib.pylab as plt

def plti(im, h=8, **kwargs):
    y = im.shape[0]
    x = im.shape[1]
    w = (y/x)*h
    plt.figure(figsize=(w,h))
    plt.imshow(im, interpolation = "none", **kwargs)
    plt.axis('off')
    
im = plt.imread('picture\sunset.jpg')
print(im.shape)
print(type(im))
plti(im)
im = im[:200,:200,:]
plti(im)
print(im.shape)
```

```python
# from Noorul8
from PIL import Image
import os

size_512 = (512,512)

for f in os.listdir("picture"):
    if f.endswith(".jpg"):
        i = Image.open(f)
        fn, fext = os.path.splitext(f)
        
        i.thumbnail(size_512)
        i.save("20Images/{}_512{}".format(fn, fext))
```
